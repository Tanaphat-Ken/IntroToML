```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
t = np.linspace(0, 2*np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

plt.figure(figsize=(8, 6))
x_left = x[x < 0]
y_left = y[x < 0]
x_right = -x_left
y_right = y_left

plt.plot(x_left, y_left, color='pink', linewidth=2)
plt.plot(x_right, y_right, color='blue', linewidth=2)

plt.title('TITLE')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')
plt.grid(True)
plt.show()
```
