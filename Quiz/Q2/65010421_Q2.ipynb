```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
theta = np.linspace(0, 2 * np.pi, 1000)
radius = 9
x = radius * np.cos(theta)
y = radius * np.sin(theta)

plt.figure(figsize=(9, 9))

# Split the circle into 4 quadrants
quarter = len(theta) // 4

# Plot each quadrant with different colors
plt.plot(x[:quarter], y[:quarter], color='red', linewidth=2, label='0° - 90°')
plt.plot(x[quarter:2*quarter], y[quarter:2*quarter], color='purple', linewidth=2, label='90° - 180°')
plt.plot(x[2*quarter:3*quarter], y[2*quarter:3*quarter], color='pink', linewidth=2, label='180° - 270°')
plt.plot(x[3*quarter:], y[3*quarter:], color='skyblue', linewidth=2, label='270° - 360°')

plt.title('Circle with 4 Colors (Radius = 9)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
url = "netflix_titles.csv"
df = pd.read_csv(url, encoding='utf-8')
x = np.arange(0, 4*np.pi, 0.1)

null_counts = df.isnull().sum()
print(null_counts)
plt.bar(null_counts.index, null_counts.values)
plt.show()

exceed = null_counts[null_counts > 800].index.tolist()
df = df.dropna(subset=exceed, inplace=False)

null_after = df.isnull().sum()
plt.bar(null_after.index, null_after.values)
plt.show()

print(df.head(9))
print("=" * 20)
print(df.tail(9))
```
