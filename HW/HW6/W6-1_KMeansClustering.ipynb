# K Mean Clustering From Scratch

```python
# First, we will create a helper function to calculate the Euclidean distance between two points.
def euclidean(point, centroid):
  return np.sqrt(np.sum((point - centroid) ** 2))
```

```python
def showplot(centroids):
  plt.figure(figsize=(9, 3.5))
  plt.plot(X[:,0], X[:,1], "k*", label="Unlabel instance")
  plt.plot([x for x, _ in centroids],[y for _, y in centroids], '+', markersize=10,)
  plt.xlabel("Petal length", fontsize=10)
  plt.xlabel("Petal width", fontsize=10)
  plt.legend(fontsize=10)
  plt.show()
```

```python
# Next, we need a way to initialize the starting point of each centroid.
# We will calculate it by uniformly distributing each centroid in the graph.
from numpy.random import uniform
import numpy as np

n_clusters = 3 # The data set have 3 class hench 3 cluster centroids we need to make

minpoint, maxpoint = np.min(X[:,:], axis=0), np.max(X[:,:], axis=0)
centroids = np.array([uniform(low=minpoint, high=maxpoint) for _ in range(n_clusters)])
print(minpoint, maxpoint)
print(centroids)
```

```python
showplot(centroids)
```

```python
def showplotwithclass(sorted_x, centroids):
    plt.figure(figsize=(9, 3.5))

    colors = plt.cm.tab20.colors

    for i, cluster_points in enumerate(sorted_x):
        if len(cluster_points) == 0:  # Skip empty clusters
            continue
        cluster_points = np.array(cluster_points)
        plt.plot(cluster_points[:, 0], cluster_points[:, 1],
                 marker='o', linestyle='', color=colors[i % len(colors)],
                 label=f"Class {i + 1} instances")

    plt.plot([x for x, _ in centroids],
             [y for _, y in centroids],
             'k+', markersize=10, label="Centroids")

    plt.xlabel("Petal length", fontsize=10)
    plt.ylabel("Petal width", fontsize=10)
    plt.legend(fontsize=10)
    plt.show()
```

```python
def calculateClass(X, centroids):
  sorted_x = [[] for _ in range(n_clusters)]  # Independent lists
  for x in X:
    dists = [euclidean(x, centroid) for centroid in centroids]
    centroid_idx = np.argmin(dists)
    sorted_x[centroid_idx].append(x)

  sorted_x = [np.array(cluster) for cluster in sorted_x]

  showplotwithclass(sorted_x, centroids)
  return sorted_x
```

```python
res = calculateClass(X, centroids)
```

```python
# After that, we use the result to adjust the current centroid.
# Reassign centroids as the mean of the points belonging to them.
def adjustCentroid(sorted_x, centroids):
  prev_centroids = centroids
  centroids = [np.mean(cluster, axis=0) for cluster in sorted_x]
  for i, centroid in enumerate(centroids):
    if np.isnan(centroid).any():  # Catch any np.nans, resulting from a centroid having no points
      centroids[i] = prev_centroids[i]

  showplotwithclass(sorted_x, centroids)
  return centroids
res = adjustCentroid(res, centroids)
```

```python
# We will create a computation loop like this.
def train(X, n_clusters):
  minpoint, maxpoint = np.min(X[:,:], axis=0), np.max(X[:,:], axis=0)
  centroids = np.array([uniform(low=minpoint, high=maxpoint) for _ in range(n_clusters)])
  while True:
    sorted_x = calculateClass(X, centroids)
    new_centroid = adjustCentroid(sorted_x, centroids)
    if np.allclose(centroids, new_centroid): break
    centroids = new_centroid
train(X, 3)
```

```python
import kagglehub
import os

# Download latest version
folder_path = kagglehub.dataset_download("sushilyeotiwad/wheat-seed-dataset")

print(os.listdir(folder_path))
df = pd.read_csv(folder_path + '/seeds_dataset.csv')
df = df.rename(columns={'Class_(1, 2, 3)': 'Class'})

class_map = {1: 'Kama', 2: 'Rosa', 3: 'Canadian'}
df['Class_Name'] = df['Class'].map(class_map)
df = df.drop(columns=['Class'])

df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from itertools import combinations
import warnings
warnings.filterwarnings('ignore')

#  Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
class_names = iris.target_names

#  DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['Class_Name'] = [class_names[i] for i in y]

print(f"Dataset Info:")
print(f" Shape: {df.shape}")
print(f" Features: {feature_names}")
print(f" Classes: {class_names}")
print(f" Samples per class: {np.bincount(y)}")

print(f"\nFirst 5 rows:")
print(df.head())

# ================================
#  1:  3 features 
# ================================
print(f"\n{'='*50}")
print(" 1:  3 Features  K-Means")
print(f"{'='*50}")

#  3 features  (sepal length, sepal width, petal length)
selected_features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']
X_selected = df[selected_features].values

print(f"Selected features: {selected_features}")
print(f"Selected data shape: {X_selected.shape}")

# ================================
#  2:  K-Means Clustering (k=3)
# ================================
print(f"\n{'='*50}")
print(" 2: K-Means Clustering (k=3)")
print(f"{'='*50}")

# Standardize 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)

#  cluster labels  dataframe
df_results = df.copy()
df_results['cluster'] = cluster_labels

print(f"K-Means completed!")
print(f"Cluster centers (scaled):")
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"  Cluster {i}: {center}")

print(f"\nCluster distribution:")
unique, counts = np.unique(cluster_labels, return_counts=True)
for cluster, count in zip(unique, counts):
    print(f"  Cluster {cluster}: {count} samples")

# ================================
#  3:  Pivot Table
# ================================
print(f"\n{'='*50}")
print(" 3: Pivot Table")
print(f"{'='*50}")

#  pivot table
pivot_table = pd.crosstab(df_results['cluster'], df_results['Class_Name'], margins=True)
print("Pivot Table - Cluster vs Class_Name:")
print(pivot_table)

# ================================
#  4: Majority Voting
# ================================
print(f"\n{'='*50}")
print(" 4: Majority Voting")
print(f"{'='*50}")

#  class  cluster
cluster_predictions = {}
for cluster in range(3):
    cluster_data = df_results[df_results['cluster'] == cluster]
    class_counts = cluster_data['Class_Name'].value_counts()
    majority_class = class_counts.index[0]
    cluster_predictions[cluster] = majority_class
    print(f"Cluster {cluster}  Predicted as '{majority_class}' ({class_counts.iloc[0]} samples)")

# ================================
#  5:  3 columns 
# ================================
print(f"\n{'='*50}")
print(" 5:  Columns ")
print(f"{'='*50}")

#  column Predict
df_results['Predict'] = df_results['cluster'].map(cluster_predictions)

#  column Status
df_results['Status'] = df_results['Class_Name'] == df_results['Predict']

print("DataFrame with new columns:")
print(df_results[['Class_Name', 'cluster', 'Predict', 'Status']].head(10))

print(f"\nStatus summary:")
status_counts = df_results['Status'].value_counts()
print(f" True (correct): {status_counts.get(True, 0)}")
print(f" False (incorrect): {status_counts.get(False, 0)}")

# ================================
#  6:  Accuracy  3D Scatter Plot
# ================================
print(f"\n{'='*50}")
print(" 6: Accuracy  3D Scatter Plot")
print(f"{'='*50}")

#  accuracy
accuracy = df_results['Status'].mean()
print(f"Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")

#  3D Scatter Plot
fig = plt.figure(figsize=(15, 6))

# Plot 1:  Class_Name 
ax1 = fig.add_subplot(121, projection='3d')
colors = ['red', 'green', 'blue']
for i, class_name in enumerate(class_names):
    mask = df_results['Class_Name'] == class_name
    ax1.scatter(df_results.loc[mask, selected_features[0]], 
                df_results.loc[mask, selected_features[1]], 
                df_results.loc[mask, selected_features[2]], 
                c=colors[i], label=f'True: {class_name}', alpha=0.7, s=50)

ax1.set_xlabel(selected_features[0])
ax1.set_ylabel(selected_features[1])
ax1.set_zlabel(selected_features[2])
ax1.set_title('True Class Labels')
ax1.legend()

# Plot 2:  Predict
ax2 = fig.add_subplot(122, projection='3d')
for i, class_name in enumerate(class_names):
    mask = df_results['Predict'] == class_name
    ax2.scatter(df_results.loc[mask, selected_features[0]], 
                df_results.loc[mask, selected_features[1]], 
                df_results.loc[mask, selected_features[2]], 
                c=colors[i], label=f'Predicted: {class_name}', alpha=0.7, s=50)

ax2.set_xlabel(selected_features[0])
ax2.set_ylabel(selected_features[1])
ax2.set_zlabel(selected_features[2])
ax2.set_title('Predicted Class Labels')
ax2.legend()

plt.tight_layout()
plt.show()

print(f"3D Scatter plots created showing:")
print(f" Left: True class labels")
print(f" Right: Predicted class labels from K-Means + Majority Voting")

# ================================
#  7:  3 features  accuracy 
# ================================
print(f"\n{'='*50}")
print(" 7:  3 Features  Accuracy ")
print(f"{'='*50}")

#  3 features  4 features
all_features = feature_names
feature_combinations = list(combinations(all_features, 3))

best_accuracy = 0
best_features = None
best_results = None

print("Testing all possible 3-feature combinations:")

for i, features in enumerate(feature_combinations):
    # 
    X_test = df[list(features)].values
    
    # Standardize
    scaler_test = StandardScaler()
    X_test_scaled = scaler_test.fit_transform(X_test)
    
    # K-Means
    kmeans_test = KMeans(n_clusters=3, random_state=42, n_init=10)
    clusters_test = kmeans_test.fit_predict(X_test_scaled)
    
    # Majority voting
    df_test = df.copy()
    df_test['cluster'] = clusters_test
    
    cluster_pred = {}
    for cluster in range(3):
        cluster_data = df_test[df_test['cluster'] == cluster]
        if len(cluster_data) > 0:
            class_counts = cluster_data['Class_Name'].value_counts()
            cluster_pred[cluster] = class_counts.index[0]
        else:
            cluster_pred[cluster] = class_names[0]  # default
    
    df_test['Predict'] = df_test['cluster'].map(cluster_pred)
    df_test['Status'] = df_test['Class_Name'] == df_test['Predict']
    
    #  accuracy
    acc = df_test['Status'].mean()
    
    print(f"  {i+1}. {features}: {acc:.4f} ({acc*100:.2f}%)")
    
    if acc > best_accuracy:
        best_accuracy = acc
        best_features = features
        best_results = df_test.copy()

print(f"\nBest feature combination:")
print(f"Features: {best_features}")
print(f"Best accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")

# ================================
#  8:  3D Scatter Plot  features 
# ================================
print(f"\n{'='*50}")
print(" 8: 3D Scatter Plot  Best Features")
print(f"{'='*50}")

#  3D plot  features 
fig = plt.figure(figsize=(15, 6))

# Plot 1:  Class_Name 
ax1 = fig.add_subplot(121, projection='3d')
colors = ['red', 'green', 'blue']
for i, class_name in enumerate(class_names):
    mask = best_results['Class_Name'] == class_name
    ax1.scatter(best_results.loc[mask, best_features[0]], 
                best_results.loc[mask, best_features[1]], 
                best_results.loc[mask, best_features[2]], 
                c=colors[i], label=f'True: {class_name}', alpha=0.7, s=50)

ax1.set_xlabel(best_features[0])
ax1.set_ylabel(best_features[1])
ax1.set_zlabel(best_features[2])
ax1.set_title(f'True Class Labels\\n(Best Features - Acc: {best_accuracy:.2%})')
ax1.legend()

# Plot 2:  Predict
ax2 = fig.add_subplot(122, projection='3d')
for i, class_name in enumerate(class_names):
    mask = best_results['Predict'] == class_name
    ax2.scatter(best_results.loc[mask, best_features[0]], 
                best_results.loc[mask, best_features[1]], 
                best_results.loc[mask, best_features[2]], 
                c=colors[i], label=f'Predicted: {class_name}', alpha=0.7, s=50)

ax2.set_xlabel(best_features[0])
ax2.set_ylabel(best_features[1])
ax2.set_zlabel(best_features[2])
ax2.set_title(f'Predicted Class Labels\\n(Best Features - Acc: {best_accuracy:.2%})')
ax2.legend()

plt.tight_layout()
plt.show()
```
